#merging json files

```python
merged_df = pd.concat([df0, df1, df2, df3, df4, df5, df6, df7, df8, df9], ignore_index=True)
```

```python
merged_df.shape
```

```python
merged_df.drop_duplicates(subset=['endTime','trackName'], keep='first', inplace=True, ignore_index=True)
```

```python
merged_df.shape
```

#dropping listened songs whose msPlayed is lower than 20000

```python
merged_df = merged_df[(merged_df['msPlayed'] > 20000)]
```

```python
merged_df.shape
```

#most listened songs

```python
first_n = merged_df['trackName'].value_counts()
new_df = pd.DataFrame({'song name':first_n.index, 'times':first_n.values})
```

```python
new_df[0:50]
```

#most listened artists

```python
first_n2 = merged_df['artistName'].value_counts()
new_df2 = pd.DataFrame({'artist name':first_n2.index, 'times':first_n2.values})
```
